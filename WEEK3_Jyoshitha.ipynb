**Pandas Dataset**

```python
#importing the pandas libraries
import pandas as pd
# Loading the file
print("BEFORE CLEANING THE DATA:   ")
df = pd.read_csv("Pandas1.csv")
df
```

```python
print(df.shape)
```

```python
# Displaying the datatypes of the attributes
print(df.info())
```

```python
# Checking for missing values
missing_values = df.isnull().sum()
# Displaying the no . of missing values
print(missing_values)
```

So we have missing values of Age is 7 and rating is 1 .


```python
#SALARY COLUMN
df0 = df['Salary'].unique()
print(df0)

print("\nData type of the 'Salary' column:")
print(df['Salary'].dtype)
```

As we can see that the data type is an object we have to convert it into numeric values .


```python
# Replacing 'k' with '000'
df['Salary'] = df['Salary'].str.replace('k', '000', regex=True)

# Replacing '$' with ' '
df['Salary'] = df['Salary'].str.replace('$', '', regex=True)

# Display the updated dataset
print(df)
```

```python
#LOCATION
df0 = df['Location'].unique()
print(df0)
```

```python
#updating the Location
df['Location'] = df['Location'].str[:-3]
df
```

```python
#ESTABLISHED
df1 = pd.DataFrame({'Established': [1999, 2002, 1988, 2002, 1999, 2008, 2009, 1999, 2019, 2020, 1999, 1984, 1943, 1954, 1955, 1944, 1946, 1988, 1999, 1987, 1980, 1934, 1935, 1932]})

# Filter out -1 values and calculate the average
average_established = df1['Established'].mean()

# Display the result
print("Average Established Year:", average_established)
```

```python
df['Established'] = df['Established'].replace(-1, 1980)
df
```

```python
#EASY APPLY
df['Easy Apply'] = df['Easy Apply'].replace('-1', 'FALSE')
df
```

```python
#RATINGS
# Scale the "Rating" column to the range [0, 10]
df['Rating']= df['Rating'].clip(0, 10)
#Replace the NaN values with 0.0
df['Rating'] = df['Rating'].fillna(0.0)
#converting the rating into round figure
df['Rating'] = df['Rating'].round()
#Converting it into integer
df['Rating'] = df['Rating'].astype(int)
# Display the DataFrame with the scaled ratings
df
```

After completing the conversions, we delete the columns containing 'NaN' values.

```python
unique_established = df['Established'].unique()
unique_age = df['Age'].unique()

print("Unique Established Years:", unique_established)
print("Unique Age Values:", unique_age)
```

```python
#AGE
current_year = 2023
# Fill missing Age values based on Established year
df['Age'].fillna(current_year - df['Established'], inplace=True)
```

```python
# After cleaning
print("AFTER CLEANING THE DATA: ")
df
```

**Chipotle dataset**



```python
# Read the TSV file into a DataFrame
df = pd.read_csv("chipotle (1).tsv", sep='\t')

# Display the DataFrame
print("Before cleaning the chipotle dataset: ")
print(df.head())
```

```python
df.isnull().sum()
```

```python
print(df.shape)
```

```python
# Displaying the datatypes of the attributes
print(df.info())
```

```python
# Filling the NaN values of choice description with No choice
df['choice_description'].fillna('No Choice', inplace=True)
df
```

```python
# Handle missing values in the Item Price column (assuming filling with 0.0)
df['item_price'].fillna('$0.0', inplace=True)
df
```

```python
# Checking If there is any dupilcate values present in the dataset
print("Duplicated Entries that are present in the dataset: ")
df.duplicated().sum()
```

```python
# Droping the duplicates
df = df.drop_duplicates()
df
```

```python
# Resetting the indices after removing the duplicates
df = df.reset_index(drop=True)
df
```

```python
df['item_name'].str.lower()
```

```python
# Display unique variations in the standardized Item Name
unique_item_names = df['item_name'].unique()
print("Standardized Item Names:\n", unique_item_names)
```

```python
# Split the 'item_price' column on '$' and convert to numeric
df['item_price'] = df['item_price'].str.replace('$', '').astype(float)

# Group by order ID and aggregate the rows
df = df.groupby('order_id').agg({
    'quantity': 'sum',
    'item_name': ', '.join,
    'choice_description': ', '.join,
    'item_price': 'sum'
}).reset_index()
print(df)
```
